# Population Estimate 2016

```python
import pandas as pd

df = pd.read_excel("C:/Users/bhagi/Unified_Mentor/uscensusbureau-big-game-census-data-visualization/All Places Census 2016 Population Estimates.xlsx")
df.head()
```

```python
df.describe()
df.info()
```

```python
df['April 1, 2010 - Census'] = df['April 1, 2010 - Census'].replace('(X)', 0)
df['April 1, 2010 - Census'].head(19310)
```

```python
df = df.dropna()
df.info()
```

```python
# List of population columns
population_columns = [
    'April 1, 2010 - Census', 'April 1, 2010 - Estimates Base', 
    'Population Estimate (as of July 1) - 2010', 'Population Estimate (as of July 1) - 2011',
    'Population Estimate (as of July 1) - 2012', 'Population Estimate (as of July 1) - 2013',
    'Population Estimate (as of July 1) - 2014', 'Population Estimate (as of July 1) - 2015',
    'Population Estimate (as of July 1) - 2016'
]

# Convert population columns to numeric types
for col in population_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Display data types to confirm conversion
print(df.dtypes)
```

```python
df['Geographic ID'] = df['Geographic ID'].astype(str)
df['Geography, full name (City, State)'] = df['Geography, full name (City, State)'].astype(str)
df.dtypes
```

```python
df = df.drop_duplicates()
df.info()
```

```python
cleaned_file_path = 'C:/Users/bhagi/Unified_Mentor/uscensusbureau-big-game-census-data-visualization/Cleaned_Census_Data_2016.xlsx'
df.to_excel(cleaned_file_path, index=False)
```
